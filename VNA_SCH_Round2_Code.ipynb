```python
SUPPLY=SUPPLY.sort_values(by=['ITEM','NGÀY HẾT HẠN TỒN KHO'])
SUPPLY
```

```python
SUPPLY['MONTH']=SUPPLY['HẠN SỬ DỤNG'].dt.month
SUPPLY
```

```python
DF_MERGE=DEMAND_FINAL.merge(SUPPLY,how='left',on=['ITEM','MONTH'])
DF_MERGE.head(12)
```

```python
DF_MERGE=DF_MERGE.fillna(0)
DF_MERGE.head()
```

```python
DF_MERGE['SỐ LƯỢNG']=DF_MERGE['SỐ LƯỢNG'].astype('int64')
```

```python
DF_MERGE['THỪA/THIẾU']=DF_MERGE['SỐ LƯỢNG']-DF_MERGE['NGLIEU_DEMAND_QTY']
DF_MERGE.head()
```

```python
del DF_MERGE['NGÀY VỀ']
del DF_MERGE['MONTH']
del DF_MERGE['ĐƠN VỊ']
del DF_MERGE['NGÀY HẾT HẠN TỒN KHO']
DF_MERGE.head()
```

```python
DF_MERGE['CUMSUM_THỪA/THIẾU']=DF_MERGE.groupby('ITEM')['THỪA/THIẾU'].cumsum()
DF_MERGE.head(2)   
```
